```python
#Sum for each year
total_per_year=df.loc[:,'2006':'2012'].sum()
print(total_per_year)
```

```python
import matplotlib.pyplot as plt

total_per_year.plot(kind='bar', color='skyblue')
plt.title('Total Crimes Against SCs (2006â€“2012)')
plt.xlabel('Year')
plt.ylabel('Number of Crimes')
plt.show()
```

```python
df['Total_2006_2012'] = df.loc[:, '2006':'2012'].sum(axis=1)
df.loc[:, 'Total_2006_2012'] = df.loc[:, '2006':'2012'].sum(axis=1)
# Make sure you're working on a fresh copy
df = df.copy()
```

```python
# Add total crimes column
df['Total_2006_2012'] = df.loc[:, '2006':'2012'].sum(axis=1)
```

```python
# Group by STATE/UT and sort
state_totals = df.groupby('STATE/UT')['Total_2006_2012'].sum().sort_values(ascending=False)
print(state_totals.head(5))
```

```python
# Plot top 5 states
state_totals.head(5).plot(kind='bar', color='orange')
plt.title('Top 5 States by Crimes Against SCs (2006â€“2012)')
plt.xlabel('State/UT')
plt.ylabel('Total Crimes')
plt.show()
```

```python
# Create pivot table: States vs Years
pivot_df = df.pivot_table(index='STATE/UT', values=['2006','2007','2008','2009','2010','2011','2012'], aggfunc='sum')

plt.figure(figsize=(12, 20))
sns.heatmap(pivot_df, cmap='YlOrRd', linewidths=0.5)
plt.title('Heatmap: Crimes Against SCs by State and Year (2006â€“2012)')
plt.xlabel('Year')
plt.ylabel('State/UT')
plt.show()
```
